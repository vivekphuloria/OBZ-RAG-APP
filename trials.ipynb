```python
from langchain_community.document_loaders import BSHTMLLoader, PyPDFLoader, TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from app.chains import get_chunk_classification_chain
from app.consts import LLM
import os
from tqdm import tqdm
from dotenv import load_dotenv
load_dotenv()

d_loader = {
    'txt' : TextLoader, 
    'pdf' : PyPDFLoader,
    'htm' :BSHTMLLoader,
    'html':BSHTMLLoader
}
folder = './rag_data'
splitter = RecursiveCharacterTextSplitter( chunk_size = 500, chunk_overlap = 100)
clf_chain = get_chunk_classification_chain(LLM)

l_docs = []
for file_name in tqdm(os.listdir(folder)):
    file_path = os.path.join(folder,file_name)

    if os.path.isdir(file_path):
        break

    file_format = file_name.split('.')[-1]
    user_name = file_name.split(' ')[0]
    
    loader = d_loader[file_format](file_path)
    docs_raw = loader.load()
    docs_split = splitter.split_documents(docs_raw)
    for doc in docs_split:
        clf = clf_chain.invoke(doc.page_content)['classification']
        doc.metadata['person_name'] = user_name
        doc.metadata['clf'] = clf

    l_docs+=docs_split
```

```python
def load_docs(folder, d_loader):
    l_docs_whole = []
    for file_name in tqdm(os.listdir(folder)):
        file_path = os.path.join(folder,file_name)

        if os.path.isdir(file_path):
            break

        file_format = file_name.split('.')[-1]
        user_name = file_name.split(' ')[0]
        
        loader = d_loader[file_format](file_path)
        docs_whole = loader.load()

        l_docs_whole += docs_whole
    return l_docs_whole

l_whole_docs = load_docs(folder, d_loader)

```

```python
metadata_filter = {"clf": {"$in":["Skills"]}}
fetch_k = 10
retriver = vectorstore.as_retriever(search_kwargs={"filter": metadata_filter, "k": fetch_k})

def get_retriver(clf, fetch_k=10, embeddings = OpenAIEmbeddings(), index_name = "obz-rag-app"):
    vectorstore = PineconeVectorStore(
        index_name=index_name,
        embedding=embeddings
    )
    retriver = vectorstore.\
        as_retriever(search_kwargs=
                     {"filter": {"clf": {"$in":clf}},
                      "k": fetch_k})
    return retriver




query = "Which all Organizations has Vivek worked in?"
q_clf_chain = get_classification_chain(llm, input_type='query')
clf = q_clf_chain.invoke(query)['classification']
retriver = get_retriver(clf)
retriver.invoke(query)
```
